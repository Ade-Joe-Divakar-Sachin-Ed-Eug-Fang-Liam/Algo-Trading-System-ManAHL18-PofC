# Skewed Normal Distribution

```python
def split_df(df, mu=20, sigma=20, minimum=5, maximum=150,verbose=False):
    ub = math.floor(len(df)/2)
    if (maximum > ub):
        maximum = ub
    while True:
        s = np.rint(np.random.normal(mu, sigma)).astype(int)
        if (minimum < s & s < maximum ):
            break         
    #print(minimum,s,maximum)
    return s

tst = pd.DataFrame(np.random.randint(0,100,size=(500, 1)), columns=list('A'))
vals = []
for i in range(100000):
    vals.append(split_df(tst))
vals = np.asarray(vals)
print(type(vals))

bins = np.arange(0, 150, 2)
plt.hist(vals,bins=bins)
```
